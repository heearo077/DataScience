# **실습1 - 파이 차트**

## 필요한 라이브러리 불러오기 

```python
# 시각화 라이브러리
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic  # 모자이크 도표를 위한 statsmodels 라이브러리

# 데이터 라이브러리 
import pandas as pd
import numpy as np
```

image.png

**pie(ratio, labels=labels)**

```python
ratio = [34.55, 31.45, 16.12, 17.88]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels)  
plt.show()
```

**파이차트의 옵션**
*   autopct: 파이 차트에 표시될 숫자의 형식
*   startangle: 조각이 시작되는 각도(디폴트는 0도, 양의 방향 x축)
*   counterclock: 조각의 방향(False는 시계 방향)
*   explode: 조각이 중심에서 떨어져 있는 정도

```python
ratio = [34.55, 31.45, 16.12, 17.88]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.2f%%', startangle=180, counterclock=True)
plt.show()
```

```python
ratio = [34.55, 31.45, 16.12, 17.88]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0, 0.20, 0]

plt.pie(ratio, labels=labels, autopct='%.2f%%', startangle=90, counterclock=False, explode=explode)  
plt.show()
```

**bar(x, y1, color='r')**

**bar(x, y2, bottom=y1, color='b')**

```python
x = np.arange(3)  # 주어진 범위와 간격에 따라 균일한 값을 갖는 배열을 반환
```

```python
x
```

```python
years = ['2018', '2019', '2020']
y1 = [100, 400, 900]  # list 형식으로 데이터 저장
y2 = [200, 300, 100]
y3 = [100, 100, 100]

plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y2, color='g')
plt.xticks(x, years)  # x축에 눈금값을 입력

plt.show()
```

```python
years = ['2018', '2019', '2020']
y1 = np.array((100, 400, 900))  # numpy 배열 형식으로 데이터 저장
y2 = np.array((200, 300, 100))
y3 = np.array((100, 100, 100))

plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y2+y1, color='g')
plt.xticks(x, years)  # x축에 눈금값을 입력

plt.show()
```

```python
iris = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/VIZ/data/iris.csv")
```

```python
plt.hist(iris['sepal.width'], bins=10)
plt.show()
```

```python
iris['sepal.width'].plot(kind='density', bw_method=0.15)  # bw_method는 분포를 추정하는데 사용되는 옵션
plt.show()
```

```python
#  붓꽃 종류별(setosa, versicolor, virginica) 꽃받침 길이 비교
setosa_sepal_length = iris[iris["variety"] == 'Setosa']['sepal.length']
versicolor_sepal_length = iris[iris["variety"] == 'Versicolor']['sepal.length']
virginica_sepal_length = iris[iris["variety"] == 'Virginica']['sepal.length']

setosa_sepal_length.plot(kind='density', bw_method=1)
versicolor_sepal_length.plot(kind='density', bw_method=1)
virginica_sepal_length.plot(kind='density', bw_method=1)

plt.xlabel('Length')
plt.ylabel('Density')

plt.show()
```
